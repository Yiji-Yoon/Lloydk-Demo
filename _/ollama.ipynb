```python
import ollama

response = ollama.chat(
    model='qwen2.5:3b',
    messages=[
        {'role': 'user', 'content': '나너무슬퍼'}
    ]
)

print(response['message']['content'])
```

```python
from sentence_transformers import SentenceTransformer

# 1. 모델 로드 (처음에만 실행하면 됩니다)
# 처음 실행 시 모델을 다운로드하느라 시간이 좀 걸립니다.
model = SentenceTransformer('Qwen/Qwen3-Embedding-0.6B')

# 2. 임베딩할 문장
sentences = ["안녕하세요", "이것은 테스트 문장입니다."]

# 3. 임베딩 실행 (문장을 벡터 숫자로 변환)
embeddings = model.encode(sentences)

# 4. 결과 확인
print(embeddings.shape) 
# 결과 예시: (2, 1024) -> 문장 2개가 각각 1024개의 숫자로 변환됨
print(embeddings[0][:5]) # 첫 번째 문장의 숫자 5개만 맛보기로 출력
```
